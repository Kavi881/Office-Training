```python
#Convert the dteday column into a datetime format and extract the day, month, and year into new columns.
Bike_share['dteday'] = pd.to_datetime(Bike_share['dteday'])


Bike_share['Year'] = Bike_share['dteday'].dt.year
Bike_share['Month'] = Bike_share['dteday'].dt.month
Bike_share['Day'] = Bike_share['dteday'].dt.day


Bike_share
```

```python
Q1 = Bike_share['rentals'].quantile(0.25)
Q3 = Bike_share['rentals'].quantile(0.75)


IQR = Q3 - Q1


lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR


outliers = Bike_share[(Bike_share['rentals'] < lower_bound) | (Bike_share['rentals'] > upper_bound)]
print(outliers)

```

```python
import matplotlib.pyplot as plt
```

```python
#Create a line plot showing the trend of rentals over the dteday.
plt.plot(Bike_share['dteday'], Bike_share['rentals'], color='orange')


plt.title('Trend of Rentals Over Time')
plt.xlabel('Date')
plt.xticks(rotation=95)
plt.grid(True)
plt.ylabel('Number of Rentals')
plt.figure(figsize=(16, 12))


plt.show()
```

```python
#Create a bar chart showing the average number of rentals for each season.
average_rentals_per_season = Bike_share.groupby('season')['rentals'].mean()


average_rentals_per_season.plot(kind='bar', color='blue')


plt.title('Average Number of Rentals per Season')
plt.xlabel('Season')
plt.ylabel('Average Rentals')


plt.show()
```

```python
#Plot a scatter plot to analyze the relationship between temp and rentals.
plt.scatter(Bike_share['temp'], Bike_share['rentals'], color='brown')


plt.title('Relationship Between Temperature and Rentals')
plt.xlabel('Temperature')
plt.ylabel('Number of Rentals')


plt.show()
```

```python
average_rentals_by_weathersit = Bike_share.groupby('weathersit')['rentals'].mean()
average_rentals_by_weathersit 
```

```python
total_rentals_by_month = Bike_share.groupby('Month')['rentals'].sum()
total_rentals_by_month
```

```python
correlation_matrix = Bike_share.corr()
correlation_with_rentals = correlation_matrix['rentals'].sort_values(ascending=False)
top_5_factors = correlation_with_rentals.drop('rentals').head(5)
```

```python
top_5_factors
```

```python
strongest_factor = top_5_factors.idxmax()  


plt.figure(figsize=(8, 5))
plt.scatter(Bike_share[strongest_factor], Bike_share['rentals'], color='blue')
plt.title(f'Relationship Between Rentals and {strongest_factor.capitalize()}')
plt.xlabel(strongest_factor.capitalize())
plt.ylabel('Number of Rentals')
plt.show()
```

```python
average_rentals_by_season = Bike_share.groupby('season')['rentals'].mean()
average_rentals_by_season 
```

```python
monthly_rentals_by_year = Bike_share.groupby(['Year', 'Month'])['rentals'].sum().unstack(level=0)
plt.figure(figsize=(10, 6))
monthly_rentals_by_year.plot(kind='line', marker='o', linewidth=2)
plt.title('Monthly Rentals for Each Year')
plt.xlabel('Month')
plt.ylabel('Total Rentals')
plt.legend(title='Year', labels=['2011', '2012']) 
plt.xticks(range(1, 13))  
plt.show()
```

**Insights from the patterns observed in the visuals.**

Insight from average_rentals_by_season-  

Season 1 (Winter) may show low rentals.

Season 2 (Spring) and Season 3 (Summer) could show higher rentals, indicating warmer weather may encourage more rentals.

Season 4 (Fall) might show a drop in rentals as temperatures start to decrease.

Insight from monthly_rentals_by_year-

The line chart may reveal that rentals are consistently higher in Spring and Summer, with a sharp decline in Winter.

Year-over-year trends could show growth or decline in rentals, possibly indicating shifts in biking culture, weather patterns, or even business strategies (e.g., increased availability of bikes).
